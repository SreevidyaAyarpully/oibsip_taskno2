### Data Visualization

```python
# checking the correlation between attributes using heatmap

sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
plt.show()
```

```python
# Checking the countplot of 'Region'

plt.figure(figsize=(7,7))
sns.countplot(y=df['Region'],hue=df['Region.1'])
plt.show()
```

```python
# Barplot- Estimated Employed V/S Region

plt.figure(figsize=(10,10))
sns.barplot(y="Region", x=" Estimated Employed",data=df)
plt.show()
```

#### The highest number of people employed is in Uttar Pradesh and the lowest is in Sikkim. The data provided for Sikkim is less compared to other states. This can also be the reason for the decreased number of employed people in that state. 

```python
# Barplot- Estimated Unemployment Rate V/S Region

plt.figure(figsize=(10,10))
sns.barplot(y="Region", x=" Estimated Unemployment Rate (%)",data=df)
plt.show()
```

#### The highest unemployment is in  Haryana, while the lowest is in Meghalaya. 

```python
# Barplot- Estimated Labour Participation Rate V/S Region

plt.figure(figsize=(10,10))
sns.barplot(y="Region", x=" Estimated Labour Participation Rate (%)",data=df)
plt.show()
```

#### The estimated labour participation rate is high in Meghalaya whereas its comparitively low in Kerala. This might be the reason why Meghalaya has lower unemployment rate compared to other states.
